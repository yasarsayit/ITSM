# [dbo].[usp_GetRequestSuggestions]

```sql
/****** Object:  StoredProcedure [dbo].[usp_GetRequestSuggestions]    Script Date: 3.05.2026 22:11:17 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE   PROCEDURE [dbo].[usp_GetRequestSuggestions]
    @SearchQuery NVARCHAR(1000)
AS
BEGIN
    SET NOCOUNT ON;

    IF @SearchQuery IS NULL OR LTRIM(RTRIM(@SearchQuery)) = ''
    BEGIN
        SELECT TOP 0
            CAST('' AS NVARCHAR(10))   AS SourceDB,
            CAST(0 AS INT)             AS RecordId,
            CAST('' AS NVARCHAR(500))  AS Title,
            CAST('' AS NVARCHAR(MAX))  AS [Description],
            CAST(NULL AS DATETIME)     AS RequestDate,
            CAST(0 AS INT)             AS RankScore;
        RETURN;
    END

    ;WITH OldResults AS
    (
        SELECT TOP 20
            SourceDB      = 'OLD',
            RecordId      = T.ID,
            Title         = ISNULL(T.BASLIK, ''),
            [Description] = ISNULL(T.ACIKLAMA, ''),
            RequestDate   = T.ACILISTARIHI,
            RankScore     = K.RANK
        FROM BILGIISLEM.dbo.TALEPLER T
         LEFT JOIN BILGIISLEM.dbo.COZUMLER C ON T.ID = C.KARTNO 
        INNER JOIN CONTAINSTABLE(BILGIISLEM.dbo.TALEPLER, (BASLIK, ACIKLAMA), @SearchQuery) K
            ON T.ID = K.[KEY]
        WHERE    (ISNULL(T.BASLIK, '') <> '' OR ISNULL(T.ACIKLAMA, '') <> '') 
        AND C.ID IS NOT NULL AND LEN(C.YORUM ) > 20
        AND YEAR(T.ACILISTARIHI) > 2022
    ),
    NewResults AS
    (
        SELECT TOP 20
            SourceDB      = 'NEW',
            RecordId      = R.RequestId,
            Title         = ISNULL(R.Topic, '') COLLATE Turkish_CI_AS, 
            [Description] = ISNULL(R.[Description], '')  COLLATE Turkish_CI_AS,
            RequestDate   = R.RequestDate,
            RankScore     = K.RANK
        FROM SECIL_ITSM_TEST.dbo.Requests R
        INNER JOIN CONTAINSTABLE(SECIL_ITSM_TEST.dbo.Requests, (Topic, [Description]), @SearchQuery) K
            ON R.RequestId = K.[KEY]
        WHERE
            ISNULL(R.isDeleted, 0) = 0
            AND (ISNULL(R.Topic, '') <> '' OR ISNULL(R.[Description], '') <> '')
    )
    SELECT TOP 5
        SourceDB,
        RecordId,
        Title,
        [Description],
        RequestDate,
        RankScore
    FROM
    (
        SELECT * FROM OldResults
        UNION ALL
        SELECT * FROM NewResults
    ) X
    ORDER BY RankScore DESC, RequestDate DESC;
END
GO
```
